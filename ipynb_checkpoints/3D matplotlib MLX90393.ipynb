```python
#MLX90393_1とCSV出力  whileで作成
import csv
import serial
import datetime
import time
import re
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
%matplotlib notebook

ser = serial.Serial('/dev/ttyACM0', timeout=2000)
smoothing = 100# log 100 data to confirm stability later


x = np.zeros(100)
y = np.zeros(100)
z = np.zeros(100)
# グラフ作成
plt.ion()
fig = plt.figure()
ax = Axes3D(fig)

while True:
    try:
        data = ser.readline(smoothing).strip().rsplit()
        listData = []
        listData.append(data)
        X = int(listData[0][0])
        Y = int(listData[0][1])
        Z = int(listData[0][2])
        x = np.append(x, X)
        y = np.append(y, Y)
        z = np.append(z, Z)
        
        #軸ラベル作成
        ax.set_xlabel("x-value")
        ax.set_ylabel("y-value")
        ax.set_zlabel("z-value")
        
        #表示範囲の設定
        ax.set_xlim(-5000, 5000)
        ax.set_ylim(-5000, 5000)
        ax.set_zlim(-5000, 5000)
        
                

        img = ax.plot( x, y, z,"o", color="r")
        plt.plot(x,y,z,linewidth=1,color='g')
        plt.show()
             
    except KeyboardInterrupt:
        ser.close()
        break
```

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3
import matplotlib.animation as animation
from IPython.display import HTML


# Attaching 3D axis to the figure
fig = plt.figure()
ax = p3.Axes3D(fig)

# Lines to plot in 3D =linspace(最初,最後,要素数)
t = np.linspace(-2 * np.pi, 2 * np.pi, 50)
#np.cos(ラジアン=弧の長さ÷半径)→−1から1まで
x1, y1, z1 = np.cos(t), np.sin(t), t / t.max()
x2, y2, z2 = t / t.max(), np.cos(t), np.sin(t)

ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
ax.set_zlim(-1.1, 1.1)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.rcParams['animation.html'] = 'html5'

# stack the plots
lns = []
for i in range(len(t)):
    ln1, = ax.plot(x1[:i], y1[:i], z1[:i], 'o-', color='steelblue')
    ln2, = ax.plot(x2[:i], y2[:i], z2[:i], 'o-', color='darkorange')
    lns.append([ln1, ln2])
    print(lns)

####line_ani = animation.ArtistAnimation(fig, lns, interval=100, blit=True)

# The empty figure shown below is unknown.
# Does anyone know how to delete this?
####line_ani




```

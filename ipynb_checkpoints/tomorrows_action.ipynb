```python
import tkinter as tk 
import random
from tkinter import ttk, PhotoImage
#グローバル変数の定義　モジュール内の全ての関数から利用する
response_area = None

#明日の予定を決める関数
def wakuwaku():
    play = ['発電勉強',
           '掃除',
           '靴磨き',
           'PYTHON勉強',
           '人口知能勉強',
           '仕事に関する勉強',
           '体幹トレーニング']
    tomorrow = random.choice(play)
    response_area.configure(text=tomorrow)
    
#画面を壁画する関数
def run():
    global response_area
    
    root = tk.Tk()
    root.title('明日の予定 : ')
    font = ('Helevetica', 14)   #フォントとフォントサイズを指定するタプルを用意
    
    canvas = tk.Canvas(
    root,
    width = 550,
    height = 200,
    relief=tk.RIDGE,
    bd=2
    )
    canvas.pack()
    
 
   
    
    #応答エリアを作成
    response_area = tk.Label(
    root,
    width=50,
    height=10,
    bg='blue',
    font=font,
    relief=tk.RIDGE,
    bd=2)
    response_area.pack()
    
    button = tk.Button(
    root,
    font=font,
    text='明日の予定はどうする？',
    command=wakuwaku
    )
    button.pack()
    root.mainloop()
    
#=================================
if __name__ == '__main__':
    run()
    
    


```
